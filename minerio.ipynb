# Todos os dados abaixo foram selecionados a coluna substância que contém ouro na sua descrição

```python
ouro_substancia= requerimento[requerimento["Substância"].str.contains("OURO")]
ouro_substancia
```

abaixo foi separado o número do processo e o ano dele e criado duas colunas que os representam, no final do dataframe, e concatenado com as demais informações

```python
ano = ouro_substancia.Processo.str.split('/', expand=True)
ano = ano.rename(columns={0: 'processo', 1:'ano'})
ano[['processo', 'ano']]=ano[['processo', 'ano']].astype(int)
```

```python
ano.head()
```

```python
concat_ouro_ano=pd.concat([ouro_substancia, ano.ano, ano.processo],axis=1)
concat_ouro_ano
```

```python
from IPython.display import display
```

# Conforme informações, para cada grama de ouro são usadas 1 grama de mercúrio, também pode ser usada a proporção de 3:1, ou seja 3 gramas de mercurio para 1 de ouro, vou criar uma coluna para estimar essa quantidade, em kg

```python
selecao_ouro['mercurio (kg)']=selecao_ouro['Quantidade Produção']
```

```python
plt.figure(figsize=(40,30))
sns.catplot(data=selecao_ouro, x='mercurio (kg)', y='UF')
plt.title("mercurio (kg) x Estado",fontsize=20)
plt.xlabel('Quantidade mercúrio kg ', fontsize=(20))
plt.ylabel('Estado', fontsize=(20))
plt.grid()
```

```python
plt.figure(figsize=(20,8))
sns.histplot(data=selecao_ouro, x="mercurio (kg)", y='UF')
plt.title("Mercúrio",fontsize=18)
plt.xlabel('Quantidade em kg de 2010 a 2021', fontsize=(20))
plt.ylabel('Estado', fontsize=(20))
plt.grid()
```

```python
selecao_mercurio_21= selecao_ouro.loc[(selecao_ouro['Ano base'] == 2021)]
selecao_mercurio_21
```
